## KNN (K - Nearest Neighbors)

#### import library

```python
# library & data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
# from sklearn.datasets import load_iris
# iris = load_iris

# column of dataset
# names = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width', 'species']

dataset = pd.read_csv('../content/sample_data/Data_all/Iris.csv')
```

```python
# training과 test dataset 분리, 시험 문제
X = dataset.iloc[:, :-1].values  # 모든 행을 사용, 열은 뒤에서 하나 뺀 값을 가져와 X에 저장
y = dataset.iloc[:, 4].values   # 모든 행을 사용, 열은 앞에서 다섯 번째 값만 가져와서 y에 저장

#X.shape
#X
#y.shape
#y
#print(X)
#print(y)
#print(X.shape, y.shape)
```

## training dataset : 80% / test dataset : 20%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20) # test dataset 20% 사용, 해당 문장 전체 시험

#X_train.shape
#X_test.shape
#y_train.shape
#y_test.shape
```

```python
# Data Standardization(표준화) : Feature의 범위가 다른 경우에만 사용
# from sklearn.preprocessing import StandardScaler
# s = StandardScaler() # 특성 스케일링 (scailing), 평균이 0, 표준편차가 1이 디도록 변환
# s.fit(X_train)
# X_train = s.transform(X_train) # 훈련 데이터를 스케일링 처리
# x_test = s.transform(X_test) # 테스트 데이터를 스케일링 처리

# 데이터 정규화 (Standardization 대신 사용)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 차원 축소 (PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# 하이퍼파라미터 튜닝을 위한 GridSearchCV 설정
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],  # k값 리스트
    'metric': ['euclidean', 'manhattan'],  # 거리 측정 방법
    'weights': ['uniform', 'distance']  # 가중치
}
```

```python
# KNN 모델 생성
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

# GridSearchCV 설정 (교차 검증: 5)
from sklearn.model_selection import GridSearchCV, cross_val_score
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='accuracy')

# 학습 수행
y_train = y_train.astype(int)
grid_search.fit(X_train_pca, y_train)

# 최적의 하이퍼파라미터 출력
print("최적의 파라미터:", grid_search.best_params_)
```

```python
# 최적의 모델을 사용한 테스트 정확도
best_knn = grid_search.best_estimator_
y_test = y_test.astype(int)  # y_test 데이터 타입을 int로 변환
y_pred = best_knn.predict(X_test_pca)

# 단일 성능 평가
print("최적의 파라미터를 사용한 정확도 값: {:.2f}".format(metrics.accuracy_score(y_test, y_pred)))
```
